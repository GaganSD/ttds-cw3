```python
import pandas as pd
import numpy as np
import warnings
import os
warnings.filterwarnings("ignore")
import json
! mkdir ~/.kaggle
! cp "kaggle.json" ~/.kaggle/
! chmod 600 kaggle.json


meta_cols = ['title', 'subtitle','keyword', 'description', 'totalDownloads', 'totalViews', 'totalVotes', 'ownerUser', 'dataset_slug']
# ownerUser and dataset_slug will be used to get URL of datasets
datasets = list()

list_page = 1
while True:
    print(f'Page {list_page} started....')
    dataset_list = ! kaggle datasets list --sort-by updated -p $list_page
    dataset_list = dataset_list[3:]
    for dataset in dataset_list:
        tmp_title = dataset.split('  ')[0]
        ! kaggle datasets metadata $tmp_title
        metadata_json = open('dataset-metadata.json', 'r')
        metadata_json = json.load(metadata_json)
        title = metadata_json['title']
        subtitle = metadata_json['subtitle']
        keywords = metadata_json['keywords']
        description = metadata_json['description']
        totalDownloads = metadata_json['totalDownloads']
        totalVotes = metadata_json['totalVotes']
        totalViews = metadata_json['totalViews']
        ownerUser = metadata_json['ownerUser']
        dataset_slug = metadata_json['datasetSlug']
        tmp_meta = [title, subtitle, keywords, description, totalDownloads, totalViews, totalVotes, ownerUser, dataset_slug]
        datasets.append(tmp_meta)
    if list_page % 5 == 0:
        datasets_df = pd.DataFrame(datasets, columns = meta_cols)
        datasets_df.to_csv(f'kaggle_dataset_df_page{list_page}.csv', index=False)
        if list_page != 5:
            os.remove(f'kaggle_dataset_df_page{list_page-5}.csv')
    list_page += 1
```
